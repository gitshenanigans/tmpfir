```python
from game_utils import *
from keras.models     import Sequential
from keras.layers     import Dense, Conv2D, BatchNormalization
from keras.optimizers import Adam, SGD
```

```python
model = Sequential()
model.add(Dense(3000, activation = 'relu', input_dim=15**2))
model.add(Dropout(0.25))
model.add(Dense(2000, activation = 'relu', input_dim=15**2))
model.add(Dropout(0.3))

model.add(Dense(15**2, activation='linear'))

opt = Adam(lr = 0.0006)
parameterList = {}
parameterList['model'] = model
parameterList['opt'] = opt

fir = FIR(CROSS, False, parameterList, 5000)
```

```python
fir.run()
```

```python
fir.test()
```
