```python
from game_utils import *
from keras.models     import Sequential
from keras.layers     import Dense, Conv2D, BatchNormalization
from keras.optimizers import Adam, SGD
```

```python
model = Sequential()
model.add(Dense(200, activation = 'relu', input_dim=15**2))
model.add(Dropout(0.3))
model.add(Dense(200, activation = 'relu', input_dim=15**2))
model.add(Dropout(0.3))

model.add(Dense(15**2, activation='linear'))

opt = SGD(lr=0.001 , momentum=0.9, decay=1e-18, nesterov=False)

parameterList = {}
parameterList['model'] = model
parameterList['opt'] = opt

fir = FIR(CROSS, False, parameterList, 3000)
```

```python
fir.run()
```

```python
fir.test()
```
